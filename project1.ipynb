```python
"""  project Title-
       Automated Transcription and Content analysis using whisper and bert"""
uploaded = files.upload()

video_file = list(uploaded.keys())[0]
print(f"Uploaded video: {video_file}")



# Output audio file name
audio_file = "audio_from_video.mp3"
subprocess.call(['ffmpeg', '-i', video_file, audio_file])  # Extract audio from video

# Step 2: Load Whisper and transcribe the audio

model = whisper.load_model("base")
result = model.transcribe(audio_file)

# Step 3: Extract transcribed text
text = result["text"]  # Extract the transcribed text from the result
print(f" Transcription: \n{text}")  # Print the transcribed text

# Step 4: Summarize the transcription

summarizer = pipeline("summarization")

# Split long text into chunks (Hugging Face models have input limits)
chunks = [text[i:i+1000] for i in range(0, len(text), 1000)]
summary = ""


for chunk in chunks:
    summary += summarizer(chunk, max_length=100, min_length=30, do_sample=False)[0]['summary_text'] + "\n"

print(" Summary:\n", summary)

kw_model = KeyBERT()
keywords = kw_model.extract_keywords(text, top_n=10)
print("\n Keywords:")
for word, score in keywords:
    print(f"{word} ({score:.2f})")
```
